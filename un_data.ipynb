## Analyzing UN Data

Examining 2022 UN data on Life Expectancy, Infant mortality, fertility and population growth

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/un_data_cleaned.csv')
```

```python
df.head()
```

```python
df_format = df.rename(columns={'Unnamed: 0':'Country'})

df_format.set_index('Country', inplace=True)

df_format.drop(columns=['Year'], inplace=True)
```

```python
df_format.head()
```

```python
life = df_format['life_expectancy']

life_sorted = life.sort_values(ascending=False)
```

```python
life_df = pd.DataFrame(life_sorted)

life_df = life_df.dropna()

life_top = life_df.head(15)
```

```python
life_sorted.describe()
```

```python
life_top.plot(kind='bar')

plt.ylim(82, 88)
plt.xticks(rotation=90)

plt.show()
```

```python
life_bottom = life_df.tail(15)

life_bottom.head()
```

```python
life_bottom.plot(kind='bar')

plt.ylim(50, 65)
plt.xticks(rotation=90)

plt.show()
```

```python
life_combined = pd.concat([life_top, life_bottom], axis=0)
```

```python
life_combined.plot(kind='bar')

plt.ylim(50, 85)
plt.xticks(rotation=90)

plt.show()
```
